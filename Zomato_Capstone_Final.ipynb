Second, we use unsupervised machine learning (clustering) to automatically group restaurants into meaningful segments based on cost, ratings, review volume, and the actual text content of customer reviews (via TF-IDF). Finally, we profile each discovered cluster in business terms and propose concrete recommendations: which restaurant segments to feature, where review-prompt notifications should be timed, and which cuisines or cost tiers carry the highest risk of negative sentiment.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# ============================================================
# 1. Clean Restaurant Metadata
# ============================================================
meta = meta_df.copy()

# Cost is stored as text with commas (e.g. "1,300") -> convert to numeric
meta['Cost'] = (meta['Cost'].astype(str)
                .str.replace(',', '', regex=False)
                .str.strip())
meta['Cost'] = pd.to_numeric(meta['Cost'], errors='coerce')

# Fill missing categorical fields
meta['Collections'] = meta['Collections'].fillna('Uncategorised')
meta['Timings']      = meta['Timings'].fillna('Not Available')

# Feature engineering: count how many cuisines / collections each restaurant has
meta['Cuisine_Count']    = meta['Cuisines'].apply(lambda x: len(str(x).split(',')) if pd.notnull(x) else 0)
meta['Collection_Count'] = meta['Collections'].apply(lambda x: len(str(x).split(',')) if x != 'Uncategorised' else 0)

# ============================================================
# 2. Clean Customer Reviews
# ============================================================
rev = rev_df.copy()

# Drop rows missing critical fields
rev.dropna(subset=['Reviewer', 'Review', 'Rating', 'Metadata', 'Time'], inplace=True)

# Rating column contains a non-numeric value "Like" -> drop those rows, convert rest to float
rev = rev[pd.to_numeric(rev['Rating'], errors='coerce').notnull()].copy()
rev['Rating'] = rev['Rating'].astype(float)

# Parse Time into a real datetime, extract useful parts
rev['Time']  = pd.to_datetime(rev['Time'], errors='coerce')
rev['Year']  = rev['Time'].dt.year
rev['Month'] = rev['Time'].dt.month
rev['Hour']  = rev['Time'].dt.hour
rev['DayOfWeek'] = rev['Time'].dt.day_name()

# Sentiment label derived from rating (used throughout EDA and as a proxy target)
def rating_to_sentiment(r):
    if r >= 4.0:
        return 'Positive'
    elif r >= 3.0:
        return 'Neutral'
    else:
        return 'Negative'

rev['Sentiment'] = rev['Rating'].apply(rating_to_sentiment)

# Review length (characters) - simple engagement signal
rev['Review_Length'] = rev['Review'].astype(str).apply(len)

# Extract reviewer's review count & follower count out of the "Metadata" text field
def parse_meta_field(text, keyword):
    try:
        match = re.search(r'(\d+)\s+' + keyword, str(text), re.IGNORECASE)
        return int(match.group(1)) if match else 0
    except Exception:
        return 0

rev['Reviewer_Reviews']   = rev['Metadata'].apply(lambda x: parse_meta_field(x, 'Review'))
rev['Reviewer_Followers'] = rev['Metadata'].apply(lambda x: parse_meta_field(x, 'Follower'))

print(f"Cleaned Metadata shape: {meta.shape}")
print(f"Cleaned Reviews  shape: {rev.shape}")
print("\nSample cleaned metadata:")
display(meta[['Name', 'Cost', 'Cuisine_Count', 'Collection_Count']].head())
print("\nSample cleaned reviews:")
display(rev[['Restaurant', 'Rating', 'Sentiment', 'Review_Length', 'Year']].head())
````

### What all manipulations have you done and insights you found?

**Manipulations performed:**
1. **Cost** was stored as text with thousand-separator commas (e.g. `"1,300"`); we stripped commas and converted it to a numeric column so it can be used in charts and clustering.
2. **Rating** was stored as text and included a non-numeric value `"Like"` (from a different review type on Zomato); rows with non-numeric ratings were removed, and the rest converted to float.
3. Rows missing critical fields (`Review`, `Rating`, `Time`, etc.) were dropped — these can't contribute to sentiment or temporal analysis.
4. **Time** was parsed into a real datetime so we can extract `Year`, `Month`, `Hour`, and `DayOfWeek` for temporal analysis.
5. A derived **Sentiment** label (Positive/Neutral/Negative) was created from the Rating, since no separate sentiment label exists in the raw data — this becomes our proxy target for NLP analysis.
6. **Review_Length** and parsed **Reviewer_Reviews**/**Reviewer_Followers** (extracted from the free-text `Metadata` field using regex) were engineered as new numeric features, since reviewer experience/engagement can affect rating behavior.
7. **Cuisine_Count** and **Collection_Count** were engineered for restaurants, capturing menu diversity and how many curated Zomato tags each restaurant has — both useful clustering features.

**Insights found:** After cleaning, we retain the vast majority of reviews (only a small fraction dropped due to the "Like" rating type and missing text), and cost values now range from budget (~₹150) to premium (~₹2800) restaurants, confirming this dataset spans a wide market segment suitable for meaningful clustering.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Based on the chart experiments above, we test three hypotheses using the cleaned review and metadata:

1. Does restaurant cost have a real relationship with average rating?
2. Do restaurants with more cuisines get significantly different ratings than restaurants with fewer cuisines?
3. Is there a significant difference in average rating between weekday and weekend reviews?

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):** There is no linear correlation between restaurant Cost and Average Rating (correlation coefficient = 0).
**Alternate Hypothesis (H1):** There is a significant linear correlation between restaurant Cost and Average Rating (correlation coefficient ≠ 0).

````python
# Perform Statistical Test to obtain P-Value
avg_by_rest = rev.groupby('Restaurant')['Rating'].mean().reset_index().rename(columns={'Rating': 'Avg_Rating'})
merged_h1 = meta.merge(avg_by_rest, left_on='Name', right_on='Restaurant', how='inner').dropna(subset=['Cost', 'Avg_Rating'])

corr_coef, p_value = stats.pearsonr(merged_h1['Cost'], merged_h1['Avg_Rating'])
print(f"Pearson correlation coefficient: {corr_coef:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print(f"\nP-value ({p_value:.4f}) < alpha ({alpha}) -> Reject Null Hypothesis. Correlation IS statistically significant.")
else:
    print(f"\nP-value ({p_value:.4f}) >= alpha ({alpha}) -> Fail to Reject Null Hypothesis. No significant correlation found.")
````

````python
# Perform Statistical Test to obtain P-Value
````

##### Which statistical test have you done to obtain P-Value?

We used the **Pearson Correlation Test**.

##### Why did you choose the specific statistical test?

Pearson's test is the standard method for measuring the strength and significance of a *linear* relationship between two continuous numerical variables (Cost and Avg_Rating), which is exactly the relationship we want to test here.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):** There is no significant difference in average rating between restaurants with few cuisines (≤3) and restaurants with many cuisines (>3).
**Alternate Hypothesis (H1):** There is a significant difference in average rating between restaurants with few cuisines (≤3) and restaurants with many cuisines (>3).

````python
# Perform Statistical Test to obtain P-Value
merged_h2 = meta.merge(avg_by_rest, left_on='Name', right_on='Restaurant', how='inner').dropna(subset=['Cuisine_Count', 'Avg_Rating'])

few_cuisines  = merged_h2[merged_h2['Cuisine_Count'] <= 3]['Avg_Rating']
many_cuisines = merged_h2[merged_h2['Cuisine_Count'] > 3]['Avg_Rating']

t_stat, p_value_h2 = stats.ttest_ind(few_cuisines, many_cuisines, equal_var=False)
print(f"Few-cuisine restaurants: n={len(few_cuisines)}, mean rating={few_cuisines.mean():.3f}")
print(f"Many-cuisine restaurants: n={len(many_cuisines)}, mean rating={many_cuisines.mean():.3f}")
print(f"\nT-statistic: {t_stat:.4f}")
print(f"P-value: {p_value_h2:.4f}")

alpha = 0.05
if p_value_h2 < alpha:
    print(f"\nP-value ({p_value_h2:.4f}) < alpha ({alpha}) -> Reject Null Hypothesis. Significant difference found.")
else:
    print(f"\nP-value ({p_value_h2:.4f}) >= alpha ({alpha}) -> Fail to Reject Null Hypothesis. No significant difference found.")
````

````python
# Perform Statistical Test to obtain P-Value
````

##### Which statistical test have you done to obtain P-Value?

We used an **Independent Samples t-test (Welch's t-test)**.

##### Why did you choose the specific statistical test?

We chose Welch's t-test because we are comparing the mean rating between two independent groups (few-cuisine vs many-cuisine restaurants) that may have unequal variances and unequal sample sizes — Welch's t-test does not assume equal variance, making it the safer choice here.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):** There is no significant difference in average rating between reviews posted on weekdays vs weekends.
**Alternate Hypothesis (H1):** There is a significant difference in average rating between reviews posted on weekdays vs weekends.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
rev['Is_Weekend'] = rev['DayOfWeek'].isin(['Saturday', 'Sunday'])

weekday_ratings = rev[~rev['Is_Weekend']]['Rating']
weekend_ratings = rev[rev['Is_Weekend']]['Rating']

t_stat_h3, p_value_h3 = stats.ttest_ind(weekday_ratings, weekend_ratings, equal_var=False)
print(f"Weekday reviews: n={len(weekday_ratings)}, mean rating={weekday_ratings.mean():.3f}")
print(f"Weekend reviews: n={len(weekend_ratings)}, mean rating={weekend_ratings.mean():.3f}")
print(f"\nT-statistic: {t_stat_h3:.4f}")
print(f"P-value: {p_value_h3:.4f}")

alpha = 0.05
if p_value_h3 < alpha:
    print(f"\nP-value ({p_value_h3:.4f}) < alpha ({alpha}) -> Reject Null Hypothesis. Significant difference found.")
else:
    print(f"\nP-value ({p_value_h3:.4f}) >= alpha ({alpha}) -> Fail to Reject Null Hypothesis. No significant difference found.")
````

##### Which statistical test have you done to obtain P-Value?

We used an **Independent Samples t-test (Welch's t-test)**.

##### Why did you choose the specific statistical test?

Same reasoning as Hypothesis 2: we're comparing mean ratings between two independent groups (weekday vs weekend reviewers) where variance and sample size may differ, so Welch's t-test is the appropriate, more robust choice over a standard Student's t-test.

### 4. Textual Data Preprocessing
(It's mandatory for textual dataset i.e., NLP, Sentiment Analysis, Text Clustering etc.)

#### 1. Expand Contraction

````python
# Expand Contraction
CONTRACTIONS = {
    "won't": "will not", "can't": "cannot", "n't": " not", "'re": " are",
    "'s": " is", "'d": " would", "'ll": " will", "'t": " not",
    "'ve": " have", "'m": " am"
}

def expand_contractions(text):
    text = str(text)
    for contraction, expansion in CONTRACTIONS.items():
        text = text.replace(contraction, expansion)
    return text

rev['Review_Expanded'] = rev['Review'].apply(expand_contractions)
print("Example - before:", rev['Review'].iloc[1][:100])
print("Example - after :", rev['Review_Expanded'].iloc[1][:100])
````

#### 2. Lower Casing

````python
# Lower Casing
rev['Review_Clean'] = rev['Review_Expanded'].str.lower()
print(rev['Review_Clean'].iloc[1][:100])
````

#### 3. Removing Punctuations

````python
# Remove Punctuations
import string
rev['Review_Clean'] = rev['Review_Clean'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
print(rev['Review_Clean'].iloc[1][:100])
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
def remove_urls_digits(text):
    text = re.sub(r'http\S+|www\S+', '', text)      # URLs
    text = re.sub(r'\w*\d\w*', '', text)            # words containing digits
    return text

rev['Review_Clean'] = rev['Review_Clean'].apply(remove_urls_digits)
print(rev['Review_Clean'].iloc[1][:100])
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove Stopwords
STOPWORDS = set([
    'the','a','an','is','it','in','on','at','to','for','of','and','or','but',
    'this','that','was','are','were','be','been','have','has','had','do','did',
    'will','would','could','should','may','might','not','no','so','with','as',
    'from','by','about','up','out','just','my','your','we','they','he','she',
    'i','me','him','her','us','them','its','our','their','what','which','who',
    'when','where','how','very','really','also','more','much','some','if','than',
    'then','there','here','all','any','each','both','few','more','most','other',
    'into','through','during','before','after','above','below','between','own',
    'same','too','even','only','over','such','again','further','once'
])

def remove_stopwords(text):
    return ' '.join([w for w in text.split() if w not in STOPWORDS])

rev['Review_Clean'] = rev['Review_Clean'].apply(remove_stopwords)
print(rev['Review_Clean'].iloc[1][:100])
````

````python
# Remove White spaces
rev['Review_Clean'] = rev['Review_Clean'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
print(rev['Review_Clean'].iloc[1][:100])
````

#### 6. Rephrase Text

````python
# Rephrase Text
# Removing very short tokens (length <= 2) and restaurant/platform-specific filler
# words that appear in nearly every review and add no discriminative signal
FILLER_WORDS = {'food','place','good','great','nice','restaurant','zomato',
                 'ordered','order','time','come','came','went','go','get','got'}

def rephrase(text):
    tokens = [t for t in text.split() if len(t) > 2 and t not in FILLER_WORDS]
    return ' '.join(tokens)

rev['Review_Clean'] = rev['Review_Clean'].apply(rephrase)
print(rev['Review_Clean'].iloc[1][:100])
````

````python
# Tokenization
rev['Tokens'] = rev['Review_Clean'].apply(lambda x: x.split())
print("Example tokens:", rev['Tokens'].iloc[1][:15])
print(f"\nAverage tokens per review: {rev['Tokens'].apply(len).mean():.1f}")
````

#### 8. Text Normalization

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
# Simple rule-based stemmer (suffix stripping) - avoids needing an external NLTK download
def simple_stem(word):
    for suffix in ['ing', 'edly', 'ies', 'ed', 'es', 's']:
        if word.endswith(suffix) and len(word) - len(suffix) >= 3:
            return word[:-len(suffix)]
    return word

def stem_tokens(tokens):
    return [simple_stem(t) for t in tokens]

rev['Tokens_Stemmed'] = rev['Tokens'].apply(stem_tokens)
rev['Cleaned_Review'] = rev['Tokens_Stemmed'].apply(lambda x: ' '.join(x))

print("Before stemming:", rev['Tokens'].iloc[1][:10])
print("After  stemming:", rev['Tokens_Stemmed'].iloc[1][:10])
````

##### Which text normalization technique have you used and why?

We used a lightweight **rule-based suffix-stripping stemmer** (similar in spirit to the Porter Stemmer) instead of full lemmatization. Stemming is faster and sufficient here because our downstream task is TF-IDF vectorization for clustering and word-frequency analysis — we care about grouping word variants (e.g., 'serving', 'served', 'serves' → 'serv') together to reduce vocabulary sparsity, not about producing grammatically valid dictionary words, which is what lemmatization would additionally guarantee at higher computational cost.

#### 9. Part of speech tagging

````python
# POS Taging
# Lightweight POS tagging using a small rule-based heuristic (no external download needed).
# This is illustrative -- for production use, spaCy or NLTK's averaged perceptron tagger
# would give far higher accuracy, but requires downloading model files.
ADJ_SUFFIXES = ('ful', 'ous', 'ive', 'able', 'al', 'ic')
ADV_SUFFIXES = ('ly',)
VERB_SUFFIXES = ('ing', 'ed')

def simple_pos_tag(word):
    if word.endswith(ADV_SUFFIXES):
        return (word, 'ADV')
    elif word.endswith(VERB_SUFFIXES):
        return (word, 'VERB')
    elif word.endswith(ADJ_SUFFIXES):
        return (word, 'ADJ')
    else:
        return (word, 'NOUN')

sample_tokens = rev['Tokens'].iloc[1][:10]
tagged = [simple_pos_tag(t) for t in sample_tokens]
print("Sample POS tags:", tagged)
````

##### Which text vectorization technique have you used and why?

We used **TF-IDF (Term Frequency-Inverse Document Frequency)** vectorization rather than simple word counts (Bag of Words). TF-IDF down-weights words that appear in almost every restaurant's reviews (like generic filler words that survived cleaning) and up-weights words that are distinctive to specific restaurants — which is exactly the signal we need for clustering restaurants by what makes their reviews unique. We capped at 200 features (`max_features=200`) to keep the feature space manageable given we only have ~100 restaurant-level documents, avoiding the curse of dimensionality in later clustering.

## ***7. ML Model Implementation***

### ML Model - 1

````python
# ML Model - 1 Implementation
# Algorithm: KMeans Clustering
# First find the optimal number of clusters (K) using Elbow Method + Silhouette Score

inertias, silhouettes = [], []
K_range = range(2, 11)
for k in K_range:
    km_test = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    lbl = km_test.fit_predict(X_combined)
    inertias.append(km_test.inertia_)
    silhouettes.append(silhouette_score(X_combined, lbl))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(list(K_range), inertias, 'o-', color='#5c6bc0')
axes[0].set_title('Elbow Method', fontweight='bold'); axes[0].set_xlabel('K'); axes[0].set_ylabel('Inertia')
axes[1].plot(list(K_range), silhouettes, 's-', color='#43a047')
axes[1].set_title('Silhouette Score vs K', fontweight='bold'); axes[1].set_xlabel('K'); axes[1].set_ylabel('Silhouette')
plt.tight_layout()
plt.show()

BEST_K = list(K_range)[silhouettes.index(max(silhouettes))]
print(f"Optimal K selected: {BEST_K}")

# Fit the Algorithm
kmeans = KMeans(n_clusters=BEST_K, random_state=RANDOM_STATE, n_init=20, max_iter=500)
km_labels = kmeans.fit_predict(X_combined)

# Predict on the model
cluster_df_ordered = cluster_df_ordered.copy()
cluster_df_ordered['KMeans_Cluster'] = km_labels
print(f"\nKMeans cluster sizes:\n{pd.Series(km_labels).value_counts().sort_index()}")
````

**KMeans** partitions restaurants into K groups by minimizing the within-cluster sum of squared distances to each cluster's centroid. It is fast, scalable, and produces compact, easy-to-interpret clusters — ideal as a baseline clustering model. It assumes roughly spherical, similarly-sized clusters, which we validate using internal metrics below.

````python
# Visualizing evaluation Metric Score chart
km_sil = silhouette_score(X_combined, km_labels)
km_db  = davies_bouldin_score(X_combined, km_labels)
km_ch  = calinski_harabasz_score(X_combined, km_labels)

pca = PCA(n_components=2, random_state=RANDOM_STATE)
X_2d = pca.fit_transform(X_combined)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sc = axes[0].scatter(X_2d[:, 0], X_2d[:, 1], c=km_labels, cmap='tab10', s=90, alpha=0.85, edgecolors='grey')
plt.colorbar(sc, ax=axes[0], label='Cluster')
axes[0].set_title(f'KMeans Clusters (K={BEST_K}) - PCA 2D View', fontweight='bold')
axes[0].set_xlabel('PCA 1'); axes[0].set_ylabel('PCA 2')

metric_names = ['Silhouette\n(higher better)', 'Davies-Bouldin\n(lower better)', 'Calinski-Harabasz\n/100 (higher better)']
metric_vals  = [km_sil, km_db, km_ch/100]
axes[1].bar(metric_names, metric_vals, color=['#43a047', '#e53935', '#1e88e5'], edgecolor='white')
axes[1].set_title('KMeans Evaluation Metrics', fontweight='bold')
for i, v in enumerate(metric_vals):
    axes[1].text(i, v + 0.02, f'{v:.3f}', ha='center', fontsize=10)
plt.tight_layout()
plt.show()

print(f"Silhouette Score  : {km_sil:.4f}")
print(f"Davies-Bouldin    : {km_db:.4f}")
print(f"Calinski-Harabasz : {km_ch:.2f}")
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
# Fit the Algorithm
from itertools import product as iproduct

configs = []
for k, init, n_init in iproduct([BEST_K-1, BEST_K, BEST_K+1], ['k-means++', 'random'], [10, 20]):
    if k < 2:
        continue
    km_t = KMeans(n_clusters=k, init=init, n_init=n_init, random_state=RANDOM_STATE, max_iter=500)
    lbl = km_t.fit_predict(X_combined)
    sil = silhouette_score(X_combined, lbl)
    configs.append({'K': k, 'Init': init, 'N_Init': n_init, 'Silhouette': round(sil, 4)})

tuning_df = pd.DataFrame(configs).sort_values('Silhouette', ascending=False)
print(tuning_df.to_string(index=False))

best_cfg = tuning_df.iloc[0]

# Predict on the model (best config)
km_best = KMeans(n_clusters=int(best_cfg['K']), init=best_cfg['Init'],
                  n_init=int(best_cfg['N_Init']), random_state=RANDOM_STATE, max_iter=500)
km_best_labels = km_best.fit_predict(X_combined)
cluster_df_ordered['KMeans_Tuned'] = km_best_labels

km_sil_tuned = silhouette_score(X_combined, km_best_labels)
km_db_tuned  = davies_bouldin_score(X_combined, km_best_labels)
km_ch_tuned  = calinski_harabasz_score(X_combined, km_best_labels)

print(f"\nBest config: K={int(best_cfg['K'])}, Init={best_cfg['Init']}, N_Init={int(best_cfg['N_Init'])}")
print(f"\nBASE  -> Silhouette={km_sil:.4f}, Davies-Bouldin={km_db:.4f}, Calinski-Harabasz={km_ch:.1f}")
print(f"TUNED -> Silhouette={km_sil_tuned:.4f}, Davies-Bouldin={km_db_tuned:.4f}, Calinski-Harabasz={km_ch_tuned:.1f}")

# Before vs After comparison chart
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
labels_cmp = ['Base', 'Tuned']
for ax, base_v, tuned_v, name, better in zip(
        axes,
        [km_sil, km_db, km_ch],
        [km_sil_tuned, km_db_tuned, km_ch_tuned],
        ['Silhouette', 'Davies-Bouldin', 'Calinski-Harabasz'],
        ['higher better', 'lower better', 'higher better']):
    bars = ax.bar(labels_cmp, [base_v, tuned_v], color=['#90a4ae', '#1e88e5'], edgecolor='white')
    ax.set_title(f'{name}\n({better})', fontweight='bold')
    for bar, v in zip(bars, [base_v, tuned_v]):
        ax.text(bar.get_x()+bar.get_width()/2, bar.get_height(), f'{v:.3f}' if v<10 else f'{v:.1f}',
                ha='center', va='bottom', fontsize=10)
plt.suptitle('KMeans: Base vs Tuned - Evaluation Metric Comparison', fontweight='bold')
plt.tight_layout()
plt.show()
````

We used a **manual Grid Search** over `n_clusters`, `init` method, and `n_init`, evaluated by Silhouette Score. We chose manual grid search over `sklearn`'s `GridSearchCV` because `GridSearchCV` is built for supervised learning (it requires a scoring function based on labeled targets) and does not natively support unsupervised clustering evaluation — so we implemented an equivalent grid search loop using our own internal clustering metric (Silhouette Score) as the optimization target.

(See above — manual grid search with Silhouette Score as the optimization objective.)

Yes, hyperparameter tuning improved the KMeans model. Comparing base vs. tuned configuration:

### ML Model - 3

````python
# ML Model - 3 Implementation
# Algorithm: DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
# Find optimal eps using k-distance graph
nbrs = NearestNeighbors(n_neighbors=5).fit(X_combined)
dists, _ = nbrs.kneighbors(X_combined)
knn_dists = np.sort(dists[:, -1])

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(knn_dists, color='#8e24aa')
ax.set_title('K-Distance Graph for DBSCAN eps Selection', fontweight='bold')
ax.set_xlabel('Points (sorted)'); ax.set_ylabel('5-NN Distance')
plt.tight_layout()
plt.show()

EPS = float(np.percentile(knn_dists, 90))   # heuristic: 90th percentile of k-distances

# Fit the Algorithm
db = DBSCAN(eps=EPS, min_samples=3)
db_labels = db.fit_predict(X_combined)

# Predict on the model
cluster_df_ordered['DBSCAN_Cluster'] = db_labels

n_clusters_db = len(set(db_labels)) - (1 if -1 in db_labels else 0)
n_noise = (db_labels == -1).sum()
print(f"DBSCAN (eps={EPS:.3f}) found {n_clusters_db} clusters and {n_noise} noise points")
````

**DBSCAN** groups points that are closely packed together (high density), marking points in low-density regions as noise/outliers. Unlike KMeans and Agglomerative, it does NOT require pre-specifying the number of clusters — it discovers this automatically from the data's density structure, and it can find arbitrarily-shaped clusters. This makes it valuable as a 'sanity check' against the other two models: if DBSCAN finds a similar grouping, we gain confidence the clusters reflect real structure rather than an artifact of forcing a fixed K.

````python
# Visualizing evaluation Metric Score chart
mask = db_labels != -1
if n_clusters_db > 1 and mask.sum() > 1:
    db_sil = silhouette_score(X_combined[mask], db_labels[mask])
    db_db  = davies_bouldin_score(X_combined[mask], db_labels[mask])
    db_ch  = calinski_harabasz_score(X_combined[mask], db_labels[mask])
else:
    db_sil, db_db, db_ch = 0, 0, 0
    print("Only one cluster found - metrics not meaningful until eps is tuned (see next cell).")

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
unique_labels = sorted(set(db_labels))
cmap_db = plt.cm.tab10(np.linspace(0, 1, max(len(unique_labels), 2)))
for lbl, color in zip(unique_labels, cmap_db):
    m = db_labels == lbl
    axes[0].scatter(X_2d[m, 0], X_2d[m, 1],
                     color='grey' if lbl == -1 else color,
                     s=20 if lbl == -1 else 70,
                     alpha=0.4 if lbl == -1 else 0.85,
                     label='Noise' if lbl == -1 else f'Cluster {lbl}', edgecolors='none')
axes[0].legend(loc='best', fontsize=8)
axes[0].set_title(f'DBSCAN Clusters (eps={EPS:.2f}) - PCA 2D', fontweight='bold')
axes[0].set_xlabel('PCA 1'); axes[0].set_ylabel('PCA 2')

metric_names = ['Silhouette\n(higher better)', 'Davies-Bouldin\n(lower better)', 'Calinski-Harabasz\n/100 (higher better)']
metric_vals  = [db_sil, db_db, db_ch/100]
axes[1].bar(metric_names, metric_vals, color=['#43a047', '#e53935', '#1e88e5'], edgecolor='white')
axes[1].set_title('DBSCAN Evaluation Metrics', fontweight='bold')
for i, v in enumerate(metric_vals):
    axes[1].text(i, v + 0.02, f'{v:.3f}', ha='center', fontsize=10)
plt.tight_layout()
plt.show()

print(f"Silhouette Score  : {db_sil:.4f}")
print(f"Davies-Bouldin    : {db_db:.4f}")
print(f"Calinski-Harabasz : {db_ch:.2f}")
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
# Fit the Algorithm (grid search over eps and min_samples)
db_results = []
for eps_pct in [70, 80, 90, 95]:
    eps_val = float(np.percentile(knn_dists, eps_pct))
    for min_samp in [3, 5, 7]:
        db_t = DBSCAN(eps=eps_val, min_samples=min_samp)
        lbl = db_t.fit_predict(X_combined)
        n_clust = len(set(lbl)) - (1 if -1 in lbl else 0)
        if n_clust > 1:
            m = lbl != -1
            if m.sum() > 1:
                sil = silhouette_score(X_combined[m], lbl[m])
                db_results.append({'Eps_Pct': eps_pct, 'Eps': round(eps_val,3),
                                    'Min_Samples': min_samp, 'N_Clusters': n_clust,
                                    'N_Noise': (lbl==-1).sum(), 'Silhouette': round(sil,4)})

db_tune_df = pd.DataFrame(db_results).sort_values('Silhouette', ascending=False)
print(db_tune_df.to_string(index=False))

if len(db_tune_df) > 0:
    best_db_cfg = db_tune_df.iloc[0]
    # Predict on the model (best config)
    db_best = DBSCAN(eps=best_db_cfg['Eps'], min_samples=int(best_db_cfg['Min_Samples']))
    db_best_labels = db_best.fit_predict(X_combined)
    cluster_df_ordered['DBSCAN_Tuned'] = db_best_labels

    mask_t = db_best_labels != -1
    db_sil_tuned = silhouette_score(X_combined[mask_t], db_best_labels[mask_t])
    db_db_tuned  = davies_bouldin_score(X_combined[mask_t], db_best_labels[mask_t])
    db_ch_tuned  = calinski_harabasz_score(X_combined[mask_t], db_best_labels[mask_t])

    print(f"\nBest config: eps={best_db_cfg['Eps']}, min_samples={int(best_db_cfg['Min_Samples'])}")
    print(f"BASE  -> Silhouette={db_sil:.4f}, Davies-Bouldin={db_db:.4f}, Calinski-Harabasz={db_ch:.1f}")
    print(f"TUNED -> Silhouette={db_sil_tuned:.4f}, Davies-Bouldin={db_db_tuned:.4f}, Calinski-Harabasz={db_ch_tuned:.1f}")

    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    for ax, base_v, tuned_v, name, better in zip(
            axes, [db_sil, db_db, db_ch], [db_sil_tuned, db_db_tuned, db_ch_tuned],
            ['Silhouette', 'Davies-Bouldin', 'Calinski-Harabasz'],
            ['higher better', 'lower better', 'higher better']):
        bars = ax.bar(['Base', 'Tuned'], [base_v, tuned_v], color=['#90a4ae', '#8e24aa'], edgecolor='white')
        ax.set_title(f'{name}\n({better})', fontweight='bold')
        for bar, v in zip(bars, [base_v, tuned_v]):
            ax.text(bar.get_x()+bar.get_width()/2, bar.get_height(), f'{v:.3f}' if v<10 else f'{v:.1f}',
                    ha='center', va='bottom', fontsize=10)
    plt.suptitle('DBSCAN: Base vs Tuned - Evaluation Metric Comparison', fontweight='bold')
    plt.tight_layout()
    plt.show()
else:
    db_sil_tuned, db_db_tuned, db_ch_tuned = db_sil, db_db, db_ch
    print("No improved configuration found within search grid; base parameters retained.")
````

##### Which hyperparameter optimization technique have you used and why?

We used a **manual Grid Search** over `eps` (derived from percentiles of the k-distance graph: 70th-95th) and `min_samples` (3, 5, 7), evaluated by Silhouette Score on non-noise points. DBSCAN is especially sensitive to `eps`, so grid-searching around the k-distance 'elbow' region (rather than guessing blindly) is the standard, recommended tuning approach for density-based clustering.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes — see the chart above. Tuning `eps` and `min_samples` typically improves the Silhouette Score and reduces the noise-point count compared to our initial single-heuristic choice (90th percentile, min_samples=3), since the grid search explicitly searches for the density threshold that produces the most internally-cohesive, well-separated clusters.

### 1. Which Evaluation metrics did you consider for a positive business impact and why?

We considered **Silhouette Score** as the primary metric for positive business impact, because it directly measures whether discovered restaurant segments are distinct enough to act on with confidence (e.g., applying different marketing strategies per segment without segments overlapping/confusing the strategy). **Davies-Bouldin** and **Calinski-Harabasz** were used as supporting/confirmatory metrics, since relying on a single metric risks being misled by that metric's specific blind spots (e.g., Silhouette can be biased toward convex clusters, so cross-checking with Calinski-Harabasz's variance-ratio view adds robustness).

### 2. Which ML model did you choose from the above created models as your final prediction model and why?

We selected **Tuned KMeans** as our final model. It achieved the best (or near-best) Silhouette Score among all three tuned models, produces clean, evenly-sized, business-interpretable clusters, and — unlike DBSCAN — does not produce a separate 'noise' category, meaning every restaurant gets assigned to an actionable segment. Compared to Agglomerative, KMeans is also faster and easier to deploy/re-fit on new restaurant data as Zomato's catalog grows.

We explain the Tuned KMeans clusters using **cluster centroid profiling** — comparing each cluster's average feature values against the dataset-wide average — which is a standard, fully transparent explainability approach for KMeans (since the centroids ARE the model's "decision boundaries"). Features like `Avg_Rating`, `Pct_Positive`, and `Cost` are typically the most distinguishing, since they show the largest cross-cluster variance.

````python
# Cluster Profiling - Model Explainability for KMeans
profile_df = cluster_df_ordered.copy()
profile_df['Cluster'] = profile_df['KMeans_Tuned']

profile = profile_df.groupby('Cluster').agg(
    Count       = ('Name', 'count'),
    Avg_Cost    = ('Cost', 'mean'),
    Avg_Rating  = ('Avg_Rating', 'mean'),
    Avg_Reviews = ('Review_Count', 'mean'),
    Pct_Pos     = ('Pct_Positive', 'mean'),
    Avg_Cuisines= ('Cuisine_Count', 'mean'),
).round(2)

print("── Cluster Profile Summary ──")
display(profile)

# Radar chart - visualizes each cluster's normalized feature profile
features_radar = ['Avg_Cost', 'Avg_Rating', 'Avg_Reviews', 'Pct_Pos', 'Avg_Cuisines']
labels_radar   = ['Cost', 'Rating', 'Reviews', '% Positive', 'Cuisines']
N = len(features_radar)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})
colors_r = plt.cm.tab10(np.linspace(0, 1, len(profile)))
for (cluster_id, row), color in zip(profile.iterrows(), colors_r):
    vals = [(row[f] - profile[f].min()) / (profile[f].max() - profile[f].min() + 1e-9) for f in features_radar]
    vals += vals[:1]
    ax.plot(angles, vals, 'o-', linewidth=2, color=color, label=f'Cluster {cluster_id}')
    ax.fill(angles, vals, alpha=0.1, color=color)
ax.set_xticks(angles[:-1]); ax.set_xticklabels(labels_radar, size=11)
ax.set_title('Cluster Radar Chart - Normalized Feature Profiles', fontweight='bold', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()

print("\nBusiness segment interpretation:")
print("  - High Cost + High Rating  -> 'Premium Dining' segment")
print("  - Low Cost + High Reviews  -> 'Popular Budget Eats' segment")
print("  - Low Rating + Low Reviews -> 'Struggling Restaurants' (needs intervention)")
print("  - Mid Cost + Diverse Cuisines -> 'Versatile Family Dining' segment")
````
